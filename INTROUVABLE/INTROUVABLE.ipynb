```python
# Sélectionner toutes les lignes où "INTROUVABLE" apparaît dans la colonne dci
introuvables = data[data["DCI"].str.contains("INTROUVABLE", na=False)]

introuvables
```

```python
import re

# Fonction pour extraire tous les mots entre parenthèses après INTROUVABLE
def extract_introuvable_meds(text):
    return re.findall(r'INTROUVABLE \((.*?)\)', text)

# Appliquer la fonction à la colonne DCI de ton dataframe introuvables
all_meds = [med for sublist in introuvables["DCI"].apply(extract_introuvable_meds) for med in sublist]

# Créer un nouveau DataFrame avec une seule colonne
df_introuvables_mots = pd.DataFrame(all_meds, columns=["Nom_introuvable"])

df_introuvables_mots
```
