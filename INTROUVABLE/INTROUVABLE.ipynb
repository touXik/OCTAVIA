```python
import pandas as pd
import numpy as np

# Chemins des fichiers Excel
fichier1 = "data/ARV_JL_MBVR.xlsx"  # Remplacez par le chemin de votre fichier
fichier2 = "data/out/INTROUVABLE.xlsx"  # Remplacez par le chemin de votre fichier

# Charger les fichiers Excel
print("Chargement des fichiers...")
df1 = pd.read_excel(fichier1)
df2 = pd.read_excel(fichier2)

print(f"Fichier 1: {df1.shape[0]} lignes, {df1.shape[1]} colonnes")
print(f"Fichier 2: {df2.shape[0]} lignes, {df2.shape[1]} colonnes")

# Extraire tous les médicaments du premier fichier (toutes les colonnes)
medicaments_df1 = set()

# Parcourir toutes les colonnes du premier DataFrame
for col in df1.columns:
    # Extraire toutes les valeurs non nulles de la colonne
    valeurs = df1[col].dropna().astype(str).str.strip()
    
    # Pour les colonnes qui contiennent plusieurs médicaments séparés par des virgules
    for valeur in valeurs:
        # Séparer par virgule et nettoyer
        meds = [m.strip() for m in str(valeur).split(',')]
        medicaments_df1.update(meds)

# Extraire tous les médicaments du deuxième fichier (colonne Nom_introuvable)
medicaments_df2 = set()

# Si la colonne s'appelle "Nom_introuvable"
if 'Nom_introuvable' in df2.columns:
    valeurs = df2['Nom_introuvable'].dropna().astype(str).str.strip()
    medicaments_df2.update(valeurs)
else:
    # Sinon, prendre la première colonne
    valeurs = df2.iloc[:, 0].dropna().astype(str).str.strip()
    medicaments_df2.update(valeurs)

print(f"\nNombre de médicaments uniques dans fichier 1: {len(medicaments_df1)}")
print(f"Nombre de médicaments uniques dans fichier 2: {len(medicaments_df2)}")

# Trouver les médicaments communs (insensible à la casse)
medicaments_df1_lower = {m.lower(): m for m in medicaments_df1 if m and m != 'nan'}
medicaments_df2_lower = {m.lower() for m in medicaments_df2 if m and m != 'nan'}

medicaments_communs = []
for med_lower in medicaments_df2_lower:
    if med_lower in medicaments_df1_lower:
        # Garder la version originale du fichier 1
        medicaments_communs.append(medicaments_df1_lower[med_lower])

# Trier par ordre alphabétique
medicaments_communs.sort()

print(f"\nNombre de médicaments communs: {len(medicaments_communs)}")
print("\nMédicaments communs:")
for med in medicaments_communs:
    print(f"  - {med}")

# Créer un DataFrame avec les résultats
df_resultat = pd.DataFrame({
    'Medicaments_Communs': medicaments_communs
})

# Exporter vers Excel
fichier_sortie = "medicaments_communs.xlsx"
df_resultat.to_excel(fichier_sortie, index=False)
print(f"\nRésultats exportés vers: {fichier_sortie}")

# Optionnel: Afficher aussi les médicaments qui sont dans df2 mais pas dans df1
medicaments_non_trouves = []
for med_lower in medicaments_df2_lower:
    if med_lower not in medicaments_df1_lower:
        # Retrouver le nom original
        for m in medicaments_df2:
            if m.lower() == med_lower:
                medicaments_non_trouves.append(m)
                break

if medicaments_non_trouves:
    print(f"\nMédicaments dans fichier 2 mais pas dans fichier 1: {len(medicaments_non_trouves)}")
    df_non_trouves = pd.DataFrame({
        'Medicaments_Non_Trouves': sorted(medicaments_non_trouves)
    })
    df_non_trouves.to_excel("medicaments_non_trouves.xlsx", index=False)
    print("Exportés vers: medicaments_non_trouves.xlsx")
```
