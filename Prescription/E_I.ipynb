```python
import unicodedata

def nettoyer_texte(x):
    """
    Nettoie une chaîne de texte :
    - convertit en str
    - supprime les espaces au début et à la fin
    - met en majuscule
    - supprime les accents
    """
    x = str(x).strip().upper()
    x = unicodedata.normalize("NFD", x).encode("ascii", "ignore").decode("utf-8")
    return x
```

```python
ei_m = ei_m[["TYPE_EI"]]
ei_m["TYPE_EI"]=ei_m["TYPE_EI"].apply(nettoyer_texte)
```
