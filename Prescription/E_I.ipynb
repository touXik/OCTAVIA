```python
import pandas as pd 
```

```python
data = pd.read_excel("df1.xlsx")
```

```python
df = data[['ID','Prelibt','CIP','DOSE','dosage','Freq','Durée','DCI_FINAL','ATC']]
```

```python
df = df.rename(columns={'DCI_FINAL': 'DCI'})
```

```python
df['is_ARV'] = df['ATC'].str.startswith('J05', na=False).astype(int)
```

```python
df
```

```python

EI = pd.read_excel("data/classe therapeutique et effets indésirables (1).xlsx", sheet_name=0)
```

```python
EI.head(10)
```

```python
arv_classes = {
    'J05AF': 'INTI',
    'J05AG': 'INNTI',
    'J05AE': 'IP',
    'J05AJ': 'INI',
    'J05AX': 'INI / INSTI'
}
```

```python
df['classe_ARV'] = df['ATC'].map(arv_classes)
```

```python
df['classe_ARV'] = df['ATC'].apply(lambda x: arv_classes.get(str(x)[:5]) if pd.notna(x) else None)
```

```python
df
```

```python
ei_dic=EI.iloc[43:,:2]
```

```python
# Suppression des espaces début / fin
df["DCI_clean"] = df["DCI"].str.strip()

# Suppression des espaces autour des ;
df["DCI_clean"] = df["DCI"].str.replace(r"\s*;\s*", ";", regex=True)
```

```python
ei_dic["DCI"]=ei_dic["DCI"].str.strip()
ei_dic["DCI"]=ei_dic["DCI"].str.replace(r"\s*;\s*",";",regex=True)
```

```python
# df : ton DataFrame
# colonne1 = clés, colonne2 = valeurs

mon_dict = dict(zip(ei_dic["DCI"], ei_dic["CLASSE TH"]))
```

```python
mon_dict
```

```python
# sélectionner seulement les ATC J05AR (5 premiers caractères)
mask = df["ATC"].str[:5] == "J05AR"

# mapping direct DCI → classe ARV
df.loc[mask, "classe_ARV"] = df.loc[mask, "DCI_clean"].map(mon_dict)
```

```python
df=df.drop("DCI_clean",axis=1)
```
