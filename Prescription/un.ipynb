# dci cip7 

```python
import pandas as pd 
```

```python
df_ref = pd.read_excel("data/bdd_c.xlsx")
df = pd.read_excel("data/out/Prescription.xlsx")
```

```python
df_ref["cip7"] = df_ref["cip7"].astype(str)

df_ref = (
    df_ref
    .assign(cip7=df_ref["cip7"].str.split(";"))
    .explode("cip7")
)
```

```python
df_ref["cip7"] = (
    df_ref["cip7"]
    .str.strip()          # enlève espaces début/fin
    .replace("", pd.NA)   # remplace vide par NA
)

df_ref = df_ref.dropna(subset=["cip7"])
```

```python
df_ref = df_ref.drop_duplicates(subset=["cip7"])
```

```python
df["CIP"] = (
    df["CIP"]
    .astype(str)
    .str.strip()
    .replace("", pd.NA)
)

# df = df.dropna(subset=["CIP"])
```

```python
df["CIP"] = df["CIP"].astype(str)
df_ref["cip7"] = df_ref["cip7"].astype(str)
```

```python
df = df.merge(
    df_ref[["cip7", "dci", "atc"]],
    left_on="CIP",
    right_on="cip7",
    how="left"
)
```

```python
import numpy as np

cols = ["dci", "atc", "DCI", "ATC"]

df[cols] = df[cols].replace("", np.nan)
```

```python
df["DCI"] = df["dci"].combine_first(df["DCI"])
df["ATC"] = df["atc"].combine_first(df["ATC"])
```

```python
df = df.drop(columns=["dci", "atc"])
```

```python
df = df.drop("Unnamed: 0", axis=1)
```

```python
df.to_excel("data/out/Prescription.xlsx",index=False)
```

# Nettoyage Dci 

```python
df["DCI"] = (
    df["DCI"]
    .astype(str)

    # Supprimer les parenthèses et leur contenu
    .str.replace(r"\s*\([^)]*\)", "", regex=True)

    # RÈGLE UNIQUE : enlève tout avant D' et s'arrête à ; ou ,
    .str.replace(r"[^;,]*D'([^;,]+)", r"\1", regex=True)

    .str.replace(r"\s{2,}", " ", regex=True)
    .str.strip()
)
```
