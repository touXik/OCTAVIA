```python
import pandas as pd 
df = pd.read_excel("data/out/Med_Hep_dci_atc_split.xlsx")
```

```python
df = df.drop("Med_Hep",axis=1)
```

```python
import pandas as pd

# liste dynamique de colonnes Med_X et DCI_MedX
med_cols = [col for col in df.columns if col.startswith("Med_")]
dci_cols = [col for col in df.columns if col.startswith("DCI_Med")]
atc_cols = [col for col in df.columns if col.startswith("ATC_Med")]

# format long pour Med
df_med = df[med_cols].melt(value_name='Med', var_name='col_med')

# format long pour DCI
df_dci = df[dci_cols].melt(value_name='DCI', var_name='col_dci')

# format long pour ATC
df_atc = df[atc_cols].melt(value_name='ATC', var_name='col_atc')

# fusion des positions pour aligner
df_long = pd.concat([df_med['Med'], df_dci['DCI'], df_atc['ATC']], axis=1)

# retirer les lignes vides
df_long = df_long.dropna(subset=['Med']).reset_index(drop=True)
```

```python
df_long
```

```python
df_introuvable = df_long[df_long['DCI'].str.contains("INTROUVABLE", na=False)]
df_introuvable
```

```python
df_introuvable = df_introuvable.drop_duplicates(subset=['Med'], keep='first')
```

```python
df_introuvable
```

```python
import numpy as np

# remplacer "INTROUVABLE" par NaN dans la colonne DCI
df_long['DCI'] = df_long['DCI'].replace(to_replace=r'(?i).*INTROUVABLE.*', value=np.nan, regex=True)

df_long['ATC'] = df_long['ATC'].replace(to_replace=r'(?i).*NON_TROUVE.*', value=np.nan, regex=True)

```

```python
# récupérer les lignes où Med n'est pas vide et DCI est vide
df_filtre = df_long[
    (df_long['Med'].notna()) & (df_long['Med'].str.strip() != "") & 
    ((df_long['DCI'].isna()) | (df_long['DCI'].str.strip() == ""))
]

# réindexer proprement
df_filtre = df_filtre.reset_index(drop=True)
```

```python
df_filtre = df_filtre.drop_duplicates(subset=['Med'], keep='first')
```

```python
import pandas as pd

# Charger votre fichier (remplacez 'votre_fichier.csv' par le nom de votre fichier)
df = pd.read_excel('filtre.xlsx')

# Fonction de nettoyage
def nettoyer_dci(valeur):
    if pd.isna(valeur):
        return valeur
    
    # Convertir en string
    valeur = str(valeur)
    
    # Supprimer les espaces au début et à la fin
    valeur = valeur.strip()
    
    # Remplacer tous les ; par /
    valeur = valeur.replace(';', '/')
    
    # Supprimer les espaces autour des /
    valeur = valeur.replace(' /', '/')
    valeur = valeur.replace('/ ', '/')
    
    # Mettre en majuscules
    valeur = valeur.upper()
    
    return valeur

# Appliquer le nettoyage à la colonne DCI
df['DCI'] = df['DCI'].apply(nettoyer_dci)

# Sauvegarder le résultat
df.to_excel('vu_nn.xlsx', index=False)

print("Nettoyage terminé !")
print("\nAperçu des données nettoyées :")
print(df.head(10))
```
